```python
bgr02_img=cv2.imread("boy.jpg")
```

```python
rgb_col_img=cv2.cvtColor(bgr02_img,cv2.COLOR_BGR2RGB)
                         
```

```python
plt.imshow(rgb_col_img)
plt.axis('on')
plt.title('color Image')
plt.show()
```

```python
cv2.imwrite('output_img.jpg',rgb_col_img)
```

```python
rgb_col_img.shape
```

```python
rgb_col_img.size
```

```python
flippe_img_hor=cv2.flip(rgb_col_img,1)
flippe_img_ver=cv2.flip(rgb_col_img,0)
flippe_img_both=cv2.flip(rgb_col_img,-1)
```

```python
plt.figure(figsize=[10,10])
plt.subplot(221); plt.imshow(flippe_img_hor)
plt.title("horizontal flip")
plt.subplot(222); plt.imshow(flippe_img_ver)
plt.title("ver flip")
plt.subplot(223); plt.imshow(flippe_img_both)
plt.title("both flip")
plt.subplot(224); plt.imshow(rgb_col_img)
plt.title("original")
```
